```python

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from difflib import SequenceMatcher
```

```python
local_path = os.path.abspath(os.getcwd())
```

```python
path = '../data/Reporte_Reclamos 2019.xls'
#path = '../data/Reporte_Reclamos 2020.xlsx'
data = pd.read_excel(path, index_col=0, header=0)  
```

```python
data = data[data['Reclamo/requerim'] == 'RECLAMO']
data.head(3)
```

```python
data.columns
```

```python
print('Cantidad de registros:', len(data))
print('Cantidad de reclamos:', len(np.unique(data['N° Reclamo'])))
print('Cantidad de clientes:', len(np.unique(data['N° Docum.'])))
```

```python
data_temp = data.groupby(['N° Docum.']).agg({ 'N° Reclamo': "count"})
data_temp = data_temp.sort_values(by=['N° Reclamo'])
plt.hist(data_temp['N° Reclamo'])
```

```python
data_temp = data.groupby(['N° Docum.']).agg({ 'N° Reclamo': "count"})
data_temp = data_temp.sort_values(by=['N° Reclamo'])
plt.hist(data_temp['N° Reclamo'], log = True)
```

```python
coll = 'Situación'
data_temp = data.groupby([coll]).agg({ 'N° Reclamo': "count"})
data_temp = data_temp.sort_values(by=['N° Reclamo']).reset_index()
fig = plt.figure(figsize = (10, 5))
plt.barh(data_temp[coll], data_temp['N° Reclamo'], color ='maroon')
```

```python
data_temp = data.groupby(['Canal Ingreso']).agg({ 'N° Reclamo': "count"})
data_temp = data_temp.sort_values(by=['N° Reclamo']).reset_index()
fig = plt.figure(figsize = (10, 5))
plt.barh(data_temp['Canal Ingreso'], data_temp['N° Reclamo'], color ='maroon')
```

```python
data_temp = data.groupby(['Cod Motivo','Motivo']).agg({ 'N° Reclamo': "count"})
data_temp = data_temp.sort_values(by=['N° Reclamo'], ascending = False).reset_index()
#fig = plt.figure(figsize = (10, 5))
#data_temp = data_temp.head(20)
#plt.barh(data_temp['Cod Motivo'], data_temp['N° Reclamo'], color ='maroon')
data_temp
```
