```python
import pandas as pd
```

```python
# Catalogo SBS codigos
catalog_sbs = pd.read_excel('data/Codigos SBS y niveles de reclamos BN.xlsx')

print(catalog_sbs['Categoria'].unique())
catalog_sbs.head()
```

```python
catalogo_sbs_motivo = catalog_sbs[catalog_sbs['Categoria'] == 'Motivo']
catalogo_sbs_motivo.to_csv("catalogo/catalogo_sbs_motivo.csv", index = False)
```

```python
catalogo_sbs_canal_incidencia = catalog_sbs[catalog_sbs['Categoria'] == 'Canal de Incidencia']
catalogo_sbs_canal_incidencia.to_csv("catalogo/catalogo_sbs_canal_incidencia.csv", index = False)
```

```python
catalogo_sbs_servicio = catalog_sbs[catalog_sbs['Categoria'] == 'Servicio']
catalogo_sbs_servicio.to_csv("catalogo/catalogo_sbs_servicio.csv", index = False)
```

```python
catalogo_sbs_producto = catalog_sbs[catalog_sbs['Categoria'] == 'Producto']
catalogo_sbs_producto.to_csv("catalogo/catalogo_sbs_producto.csv", index = False)
```

```python
# Catalogo de servicios
catalog_aisr = pd.read_excel('data/Analogias_de_servicio_y_motivo_AISR.xlsx', header = 1)

print(catalog_aisr.columns)
catalog_aisr.head()
```

```python
objeto_df = catalog_aisr.groupby(['Cod. Servicio', 'COD. CANAL OK',
       'DESCRIPCIÓN DE CANAL/DE INCIDENCIA OK', 'COD_SERVICIO OK',
       'SERVICIO_OK', 'Servicio',]).agg({ 'N°': "count"})
objeto_df = objeto_df.sort_values(by=['N°'], ascending = False).reset_index()
objeto_df.to_csv("catalogo/catalogo_servicios.csv", index = False)
```

```python
objeto_df = catalog_aisr.groupby(['Cod Motivo', 'COD_MOTIVO OK', 'MOTIVO_OK',
       'Motivo']).agg({ 'N°': "count"})
objeto_df = objeto_df.sort_values(by=['N°'], ascending = False).reset_index()
objeto_df.to_csv("catalogo/catalogo_motivos.csv", index = False)
```
